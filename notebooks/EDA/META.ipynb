Checking for outliers

```python
def outliers(c):
  z_scores = (data[c] - data[c].mean()) / data[c].std()
  outlier = data[np.abs(z_scores) > 3]
  print(c," outliers\n",outlier[c], "\n\n")
```

```python
outliers("Open")
outliers("High")
outliers("Low")
outliers("Close")
outliers("Volume")
```
